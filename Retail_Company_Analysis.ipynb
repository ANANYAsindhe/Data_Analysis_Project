```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

customers = pd.read_csv('/customer.csv')
purchases = pd.read_csv('/purchase.csv')
```

```python
print("Customers DataFrame Info:")
display(customers.head())
display(customers.info())
display(customers.describe())

print("\nPurchases DataFrame Info:")
display(purchases.head())
display(purchases.info())
display(purchases.describe())
```

# Task
Analyze the provided customer and purchase data to understand customer demographics and purchase patterns. Use the files "/customer.csv" and "/purchase.csv".

## Analyze customer demographics

### Subtask:
Explore the distribution of age, gender, country, and income of the customers using visualizations and summary statistics.


**Reasoning**:
Generate summary statistics and value counts for the specified columns and create visualizations to explore the distributions of age, income, gender, and country as requested in the instructions.



```python
print("Summary Statistics for Age and Income:")
display(customers[['age', 'income']].describe())

print("\nValue Counts for Gender:")
display(customers['gender'].value_counts())

print("\nValue Counts for Country:")
display(customers['country'].value_counts())

plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
sns.histplot(customers['age'], kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')

plt.subplot(2, 2, 2)
sns.histplot(customers['income'], kde=True)
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Distribution of Income')

plt.subplot(2, 2, 3)
sns.countplot(x='gender', data=customers)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender')

plt.subplot(2, 2, 4)
sns.countplot(x='country', data=customers)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Distribution of Country')

plt.tight_layout()
plt.show()
```

## Analyze purchase patterns

### Subtask:
Investigate the types of products purchased, the total purchase amounts, and the frequency of purchases.


**Reasoning**:
Calculate the total price for each purchase, find the number of unique products and top 10 most frequent products, and calculate the number of purchases per customer.



```python
purchases['total_price'] = purchases['price'] * purchases['quantity']

unique_products_count = purchases['product_name'].nunique()
top_10_products = purchases['product_name'].value_counts().head(10)

purchases_per_customer = purchases['customer_id'].value_counts()

print(f"Number of unique products purchased: {unique_products_count}")
print("\nTop 10 most frequently purchased products:")
display(top_10_products)

print("\nTotal number of purchases made by each customer (first 10):")
display(purchases_per_customer.head(10))
```

**Reasoning**:
Visualize the distribution of total purchase amounts and the frequency of purchases per customer using histograms.



```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(purchases['total_price'], kde=True)
plt.xlabel('Total Purchase Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Total Purchase Amounts')

plt.subplot(1, 2, 2)
sns.histplot(purchases_per_customer, kde=True)
plt.xlabel('Number of Purchases')
plt.ylabel('Frequency')
plt.title('Distribution of Purchase Frequency per Customer')

plt.tight_layout()
plt.show()
```

## Combine data

### Subtask:
Merge the customer and purchase dataframes to analyze the relationship between customer demographics and purchase behavior.


**Reasoning**:
Merge the customers and purchases dataframes and display the first few rows and info of the merged dataframe.



```python
merged_data = pd.merge(customers, purchases, on='customer_id')
print("Merged DataFrame Info:")
display(merged_data.head())
display(merged_data.info())
```

## Visualize findings

### Subtask:
Create visualizations to present the key findings from the analysis.


**Reasoning**:
Create visualizations to show the relationship between customer age and total purchase amount, average total purchase amount by country and gender, distribution of total purchase amount by top 5 product categories, and average income by country.



```python
plt.figure(figsize=(18, 15))

# 1. Scatter plot: Age vs. Total Purchase Amount
plt.subplot(3, 2, 1)
sns.scatterplot(x='age', y='total_price', data=merged_data)
plt.xlabel('Customer Age')
plt.ylabel('Total Purchase Amount')
plt.title('Relationship between Age and Total Purchase Amount')

# 2. Bar plot: Average Total Purchase Amount by Country
plt.subplot(3, 2, 2)
avg_purchase_by_country = merged_data.groupby('country')['total_price'].mean().sort_values(ascending=False)
sns.barplot(x=avg_purchase_by_country.index, y=avg_purchase_by_country.values)
plt.xlabel('Country')
plt.ylabel('Average Total Purchase Amount')
plt.title('Average Total Purchase Amount by Country')
plt.xticks(rotation=45, ha='right')

# 3. Bar plot: Average Total Purchase Amount by Gender
plt.subplot(3, 2, 3)
avg_purchase_by_gender = merged_data.groupby('gender')['total_price'].mean().sort_values(ascending=False)
sns.barplot(x=avg_purchase_by_gender.index, y=avg_purchase_by_gender.values)
plt.xlabel('Gender')
plt.ylabel('Average Total Purchase Amount')
plt.title('Average Total Purchase Amount by Gender')

# 4. Box plot: Distribution of Total Purchase Amount by Top 5 Product Categories
plt.subplot(3, 2, 4)
top_5_products = purchases['product_name'].value_counts().head(5).index
merged_data_top_products = merged_data[merged_data['product_name'].isin(top_5_products)]
sns.boxplot(x='product_name', y='total_price', data=merged_data_top_products)
plt.xlabel('Product Category')
plt.ylabel('Total Purchase Amount')
plt.title('Distribution of Total Purchase Amount by Top 5 Product Categories')
plt.xticks(rotation=45, ha='right')

# 5. Bar plot: Average Income by Country
plt.subplot(3, 2, 5)
avg_income_by_country = customers.groupby('country')['income'].mean().sort_values(ascending=False)
sns.barplot(x=avg_income_by_country.index, y=avg_income_by_country.values)
plt.xlabel('Country')
plt.ylabel('Average Income')
plt.title('Average Income by Country')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
